```python
# Question: Find the union of two sets
print("SOLVED EXAMPLE:")
print("Find the union of two sets")
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
union_set = set1.union(set2)
print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
print(f"Union: {union_set}")
print("-" * 50)
```

```python
# Question 8: Create a set of vowels from string "hello world"

word = "hello world"
vowels = {'a', 'e', 'i', 'o', 'u'}

result = set()

for char in word:
    if char in vowels:
        result.add(char)

print(result)
```
